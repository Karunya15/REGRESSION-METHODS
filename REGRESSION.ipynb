**LINEAR ANS POLYNOMIAL REGRESSION**

```python
import pandas as pd
dataset = pd.read_csv('../input/height/HeightVsWeight.csv')
X = dataset.iloc[:, 0:1].values
y = dataset.iloc[:, 1].values

dataset.head()
```

```python

y
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X,y, color="red")
plt.plot(X, lin_reg.predict(X))
plt.title("Linear Regression")
plt.xlabel("Age")
plt.ylabel("Height")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=5)
X_poly = poly_reg.fit_transform(X)
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
plt.scatter(X,y, color="red")
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)))
plt.title("Poly Regression - Degree 5")
plt.xlabel("Age")
plt.ylabel("Height")
plt.show()
```

```python
lin_reg.predict([[45]])
```

```python
lin_reg_2.predict(poly_reg.fit_transform([[45]]))
```

DECISION TREE REGRESSOR****

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
X = dataset.iloc[:, 0:1].values
y = dataset.iloc[:, 1].values

dataset.head()
```

```python
tree_regressor = DecisionTreeRegressor(random_state = 0)
tree_regressor.fit(X, y)
```

```python
import numpy as np
X_grid = np.arange(min(X), max(X), 0.09)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.scatter(X, y, color='red', label='Actual observation points')
plt.plot(X_grid, tree_regressor.predict(X_grid), label='Tree regressor')

plt.xlabel('Age')
plt.ylabel('Height')

plt.legend()
plt.show()
```

```python
tree_regressor.predict([[45]])
```
